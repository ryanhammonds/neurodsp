```python
import numpy as np
import matplotlib.pyplot as plt

from neurodsp.sim import sim_combined
from neurodsp.filt import filter_signal
from neurodsp.timefrequency import amp_by_time
from neurodsp.plts import plot_time_series

np.random.seed(0)

n_seconds = 5
n_seconds_filter = 2
fs = 1000
freq = 10
f_alpha = (8, 12)
times = np.arange(0, n_seconds, 1/fs)

sim_components = {'sim_powerlaw' : {'exponent' : -2}, 'sim_bursty_oscillation' : {'freq' : 10}}
sig = sim_combined(n_seconds, fs, sim_components)

# Compute amplitude
sig_filt = filter_signal(sig, fs, 'bandpass', f_alpha, n_seconds=n_seconds_filter)
amp = amp_by_time(sig, fs, f_alpha, n_seconds=n_seconds_filter)

# Plot with different line styles
plot_time_series(times, [sig_filt, amp], colors=['k', 'r'],
                 ylabel='Oscillation amplitude', lw=[1, 5])

plot_time_series(times, [sig_filt, amp], colors=['k', 'r'],
                 ylabel='Oscillation amplitude', lw=[5, 1], alpha=[1, 0.5])
```
